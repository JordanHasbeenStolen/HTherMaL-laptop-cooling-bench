```python
import numpy as np
import pandas as pd

def compute_metrics(condition, warmup_cut=10, tail_seconds=60, threshold=90):
    data = cooling_data[cooling_data["condition"] == condition].copy()
    
    last = data["seconds"].max()
    tail = data[data["seconds"] >= last - tail_seconds]
    
    data_trim = data[data["seconds"] >= warmup_cut]
    
    # plateau
    plateau_mean = tail["Temp:PackageId0,0"].mean()
    plateau_median = tail["Temp:PackageId0,0"].median()
    plateau_std = tail["Temp:PackageId0,0"].std()
    
    # peak
    peak_max = data_trim["Temp:PackageId0,0"].max()
    
    # time-to-threshold
    hit = data_trim[data_trim["Temp:PackageId0,0"] >= threshold]
    time_to_threshold = None if len(hit) == 0 else hit["seconds"].min()
    
    # AUC
    x = data_trim["seconds"].to_numpy()
    y = data_trim["Temp:PackageId0,0"].to_numpy()
    auc = np.trapz(y, x)
    
    return {
        "condition": condition,
        "plateau_mean_last60": plateau_mean,
        "plateau_median_last60": plateau_median,
        "plateau_std_last60": plateau_std,
        "peak_max_after_warmup": peak_max,
        f"time_to_{threshold}C": time_to_threshold,
        "auc_after_warmup": auc,
    }

results = pd.DataFrame([
    compute_metrics("desk"),
    compute_metrics("raised"),
    compute_metrics("fan"),
])

results
```

```python
# cooling_eval.py
# Purpose:
#   Compare 3 cooling modes (desk / raised / fan) from CSV telemetry
#   using:
#     - elapsed time per condition (seconds)
#     - rolling median smoothing (window configurable)
#     - simple metrics table (plateau, peak, time-to-threshold, variability, AUC)
#
# Assumptions about CSV:
#   - has column "Time" like "2026-02-27_18:07:13" (underscore between date and time)
#   - has column "Temp:PackageId0,0"
#
# Run:
#   python cooling_eval.py

from __future__ import annotations

import sys
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# =========================
# (A) CONFIG — CHANGE THESE
# =========================

# --- Your 3 input CSV files ---
FILES = {
    "desk":   r"20260227_desk_5min.csv",
    "raised": r"20260227_rised_5min.csv",   # note: your filename uses "rised"
    "fan":    r"20260227_fan_5min.csv",
}

# --- Column names in CSV ---
TIME_COL = "Time"
TEMP_COL = "Temp:PackageId0,0"

# --- Warm-up trimming (ignore first N seconds for some metrics) ---
WARMUP_CUT_S = 10

# --- Plateau window (last N seconds of each test) ---
PLATEAU_LAST_S = 60

# --- Time-to-threshold (pick a temperature that means "already hot") ---
THRESHOLD_C = 90

# --- Smoothing for visualization ---
# IMPORTANT:
#   Your sampling step is ~2 seconds.
#   rolling_window = 50  => about 100 seconds smoothing.
ROLLING_MEDIAN_WINDOW = 50  # <<<<< CHANGE THIS (3 / 5 / 50 etc.)

# --- Plot sizing ---
FIGSIZE = (16, 5)  # wide timeline look


# =========================
# (B) LOADING + PREP
# =========================

def load_one_csv(path: str | Path, condition: str) -> pd.DataFrame:
    """
    Load one CSV and attach condition label.
    Also parse Time into pandas datetime.
    """
    path = Path(path)
    if not path.exists():
        raise FileNotFoundError(f"CSV not found: {path.resolve()}")

    df = pd.read_csv(path)
    if TIME_COL not in df.columns:
        raise KeyError(f"Missing column '{TIME_COL}' in {path.name}")
    if TEMP_COL not in df.columns:
        raise KeyError(f"Missing column '{TEMP_COL}' in {path.name}")

    df["condition"] = condition

    # Time parsing:
    # Example: "2026-02-27_18:07:13" -> "2026-02-27 18:07:13"
    df[TIME_COL] = pd.to_datetime(df[TIME_COL].astype(str).str.replace("_", " ", regex=False))

    return df


def add_seconds_elapsed(df: pd.DataFrame) -> pd.DataFrame:
    """
    Create 'seconds' = elapsed time since start of *each condition*.
    Concept:
      seconds = Time - min(Time within that condition)
    """
    start_time = df.groupby("condition")[TIME_COL].transform("min")
    df["seconds"] = (df[TIME_COL] - start_time).dt.total_seconds()
    return df


# =========================
# (C) METRICS
# =========================

def compute_metrics_for_condition(data: pd.DataFrame) -> dict:
    """
    Compute a small set of metrics for one condition.

    Metrics:
      - plateau_mean_lastN: mean temp over last PLATEAU_LAST_S seconds
      - plateau_std_lastN:  std  temp over last PLATEAU_LAST_S seconds (variability)
      - peak_max_after_warmup: max temp after WARMUP_CUT_S seconds
      - time_to_threshold: first second when temp >= THRESHOLD_C (after warmup)
      - auc_after_warmup: area under temp curve after warmup (°C * sec)
    """
    condition = str(data["condition"].iloc[0])

    # Sort by time just in case (plotting/metrics assume time order)
    data = data.sort_values("seconds")

    # Tail / plateau
    last_sec = float(data["seconds"].max())
    tail = data[data["seconds"] >= last_sec - PLATEAU_LAST_S]
    plateau_mean = float(tail[TEMP_COL].mean())
    plateau_std = float(tail[TEMP_COL].std())

    # Warmup-trimmed segment
    trimmed = data[data["seconds"] >= WARMUP_CUT_S]

    peak_max = float(trimmed[TEMP_COL].max())

    # Time to threshold
    hit = trimmed[trimmed[TEMP_COL] >= THRESHOLD_C]
    time_to_threshold = None if hit.empty else float(hit["seconds"].min())

    # AUC using trapezoid rule (NumPy modern API)
    x = trimmed["seconds"].to_numpy()
    y = trimmed[TEMP_COL].to_numpy()
    auc = None
    if len(x) >= 2:
        auc = float(np.trapezoid(y, x))

    return {
        "condition": condition,
        f"plateau_mean_last{PLATEAU_LAST_S}s_C": plateau_mean,
        f"plateau_std_last{PLATEAU_LAST_S}s_C": plateau_std,
        f"peak_max_after{WARMUP_CUT_S}s_C": peak_max,
        f"time_to_{THRESHOLD_C}C_s": time_to_threshold,
        f"auc_after{WARMUP_CUT_S}s_Cs": auc,
        "duration_total_s": float(last_sec),
    }


# =========================
# (D) PLOTTING
# =========================

def add_rolling_median_column(data: pd.DataFrame, window: int) -> pd.DataFrame:
    """
    Add a smoothed temperature column using rolling median.

    Why rolling median:
      - It "cuts spikes" better than rolling mean.
      - Large window gives "overview mode" (trend),
        small window gives "detail mode".
    """
    data = data.sort_values("seconds").copy()
    sm_col = f"temp_med{window}"
    data[sm_col] = data[TEMP_COL].rolling(window=window).median()
    return data


def plot_three_conditions(df: pd.DataFrame, window: int) -> None:
    """
    One figure, 3 lines, smoothed temp.
    """
    fig, ax = plt.subplots(figsize=FIGSIZE)

    # Style choices: you can change these freely
    styles = {
        "desk":   {"linestyle": "-",  "label": "desk"},
        "raised": {"linestyle": "--", "label": "raised"},
        "fan":    {"linestyle": ":",  "label": "fan"},
    }

    sm_col = f"temp_med{window}"

    for condition, sub in df.groupby("condition"):
        sub = add_rolling_median_column(sub, window=window)

        st = styles.get(condition, {"linestyle": "-", "label": condition})

        ax.plot(
            sub["seconds"],
            sub[sm_col],
            linestyle=st["linestyle"],
            label=st["label"],
        )

    ax.set_xlabel("seconds")
    ax.set_ylabel(f"{TEMP_COL} (rolling median window={window})")
    ax.set_title(f"CPU temp comparison (median{window})")
    ax.legend()
    ax.grid(True, alpha=0.2)

    plt.show()


# =========================
# (E) MAIN
# =========================

def main() -> int:
    # 1) Load 3 CSVs and stack rows (vertical concat)
    frames = []
    for cond, path in FILES.items():
        frames.append(load_one_csv(path, condition=cond))
    cooling_data = pd.concat(frames, ignore_index=True)

    # 2) Add elapsed seconds per condition
    cooling_data = add_seconds_elapsed(cooling_data)

    # 3) Plot (overview)
    plot_three_conditions(cooling_data, window=ROLLING_MEDIAN_WINDOW)

    # 4) Metrics table (per condition)
    rows = []
    for cond, sub in cooling_data.groupby("condition"):
        rows.append(compute_metrics_for_condition(sub))

    results = pd.DataFrame(rows).sort_values("condition").reset_index(drop=True)

    # Pretty print
    pd.set_option("display.width", 200)
    pd.set_option("display.max_columns", 50)
    print("\n=== RESULTS (per condition) ===")
    print(results.to_string(index=False))

    return 0


if __name__ == "__main__":
    try:
        raise SystemExit(main())
    except Exception as e:
        print(f"\nERROR: {e}", file=sys.stderr)
        raise
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) What we take from the overall table
# IMPORTANT: we assume cooling_data already exists and contains:
# - condition (desk/raised/fan)
# - seconds
# - Temp:PackageId0,0
temp_col = "Temp:PackageId0,0"

# 2) Normalize time so the points line up for subtraction
# seconds_int is "seconds rounded to whole numbers", so 1.999999 -> 2
data = cooling_data.copy()
data["seconds_int"] = data["seconds"].round(0).astype(int)

# 3) Make a "wide" table temporarily, ONLY FOR THIS CHART
# Before (long): seconds_int, condition, temp
# After (wide): seconds_int, desk, raised, fan
wide = (
    data.pivot_table(
        index="seconds_int",      # X-axis
        columns="condition",      # series
        values=temp_col,          # temperature
        aggfunc="mean"            # in case 2 points landed on the same second
    )
    .sort_index()
)

# 4) Choose the baseline
baseline = "raised"

# 5) Compute the difference (delta) relative to the baseline
delta = wide.sub(wide[baseline], axis=0)

# 6) Draw only two lines: desk-baseline and fan-baseline
fig, ax = plt.subplots(figsize=(12, 4))

ax.plot(delta.index, delta["desk"], label="desk - raised")
ax.plot(delta.index, delta["fan"], label="fan - raised")

# zero line: "no difference"
ax.axhline(0, linestyle="--")

ax.set_xlabel("seconds")
ax.set_ylabel("ΔTemp (°C)")
ax.set_title("Delta plot: temperature difference vs baseline (raised)")
ax.legend()

plt.show()
```

```python
# =========================
# SETTINGS (tweak here)
# =========================
TEMP_COL = "Temp:PackageId0,0"
BASELINE = "raised"

# Smoothing before the delta (so delta isn't "double noise")
# 1 = no smoothing, 3 = light, 5 = stronger
ROLLING_MEDIAN_WINDOW = 3

# If you want binning instead of an "honest intersection":
# set BIN_SECONDS = None (means we do NOT bin)
# or BIN_SECONDS = 2 / 5 / 10 (means we group time into bins)
BIN_SECONDS = None  # e.g.: 5

# =========================
# 1) Take the data and prepare a working copy
# =========================
data = cooling_data.copy()

# Important: sort so that rolling works "by time"
data = data.sort_values(["condition", "seconds"])

# =========================
# 2) (optional) time binning
# =========================
# Idea: if BIN_SECONDS is set, we use seconds_bin instead of seconds:
# seconds_bin = 0, 5, 10, 15...
if BIN_SECONDS is None:
    time_col = "seconds"
else:
    data["seconds_bin"] = (data["seconds"] // BIN_SECONDS) * BIN_SECONDS
    time_col = "seconds_bin"

# =========================
# 3) Smooth the temperature within each condition
# =========================
# Idea in plain terms:
# "for each row, take the median over the last N points"
# This reduces noise -> delta becomes more readable
if ROLLING_MEDIAN_WINDOW == 1:
    data["temp_smooth"] = data[TEMP_COL]
else:
    data["temp_smooth"] = (
        data.groupby("condition")[TEMP_COL]
            .rolling(window=ROLLING_MEDIAN_WINDOW)
            .median()
            .reset_index(level=0, drop=True)
    )

# =========================
# 4) Make a "wide" table only for computing the delta
# =========================
# Row = moment in time (seconds or seconds_bin)
# Column = condition
# Value = temp_smooth (smoothed temperature)
wide = (
    data.pivot_table(
        index=time_col,
        columns="condition",
        values="temp_smooth",
        aggfunc="mean"  # if a bin contains several points, take the mean
    )
    .sort_index()
)

# =========================
# 5) THE MAIN POINT: fix "choppiness" honestly
# =========================
# We keep only the moments in time where both the baseline and the compared condition exist.
# Otherwise delta = NaN and the line breaks up.
# Here we simply ask pandas: "drop rows where baseline is missing"
wide = wide.dropna(subset=[BASELINE])

# =========================
# 6) Compute delta relative to the baseline
# =========================
# delta_table[cond] = wide[cond] - wide[BASELINE]
delta_table = wide.sub(wide[BASELINE], axis=0)

# =========================
# 7) Draw delta (excluding baseline)
# =========================
fig, ax = plt.subplots(figsize=(12, 4))

for cond in ["desk", "fan"]:
    # Drop points where this cond has no data -> so the line stays honest
    # (you can dropna or leave gaps — your choice)
    series = delta_table[cond].dropna()
    ax.plot(series.index, series.values, label=f"{cond} - {BASELINE}")

ax.axhline(0, linestyle="--")
ax.set_xlabel(time_col)
ax.set_ylabel("ΔTemp (°C)")
title_bin = "" if BIN_SECONDS is None else f" (binned {BIN_SECONDS}s)"
title_roll = "" if ROLLING_MEDIAN_WINDOW == 1 else f" (median{ROLLING_MEDIAN_WINDOW})"
ax.set_title(f"Delta plot vs baseline '{BASELINE}'{title_bin}{title_roll}")
ax.legend()
plt.show()

# =========================
# 8) A small diagnostic (why it might have been choppy)
# =========================
print("Missing points count (after pivot):")
print(wide.isna().sum())
```

Min/Max band + mean over time

```python
# ====== SETTINGS (tweak here) ======
TEMP_COL = "Temp:PackageId0,0"
BIN_SECONDS = 5          # 2 / 5 / 10 — the bigger, the "calmer" the chart
PLOT_CONDITIONS = ["desk", "raised", "fan"]

# ====== 1) Prepare the time bin ======
data = cooling_data.copy()
data["tbin"] = (data["seconds"] // BIN_SECONDS) * BIN_SECONDS

# ====== 2) Compute min/mean/max for each condition and each bin ======
agg = (
    data.groupby(["condition", "tbin"])[TEMP_COL]
        .agg(tmin="min", tmean="mean", tmax="max")
        .reset_index()
)

# ====== 3) Draw: for each condition — mean line + min/max corridor ======
fig, ax = plt.subplots(figsize=(12, 4))

for cond in PLOT_CONDITIONS:
    sub = agg[agg["condition"] == cond].sort_values("tbin")

    # mean line
    ax.plot(sub["tbin"], sub["tmean"], label=f"{cond} mean")

    # min..max corridor (band)
    ax.fill_between(sub["tbin"], sub["tmin"], sub["tmax"], alpha=0.5)

ax.set_xlabel(f"seconds (binned every {BIN_SECONDS}s)")
ax.set_ylabel(f"{TEMP_COL} (°C)")
ax.set_title("Min/Max band (stability) + mean (trend)")
ax.legend()
plt.show()
```

Time-to-Threshold

```python
# ====== SETTINGS (tweak here) ======
TEMP_COL = "Temp:PackageId0,0"
THRESHOLD_C = 90          # temperature threshold
SMOOTH_MEDIAN_WINDOW = 3  # 1 = no smoothing; 3/5 = to remove "jitter"/noise
CONDITIONS = ["desk", "raised", "fan"]

data = cooling_data.copy().sort_values(["condition", "seconds"])

# 1) (optional) smooth the temperature within each condition,
# so that single noise spikes don't produce a "false early threshold crossing"
if SMOOTH_MEDIAN_WINDOW == 1:
    data["temp_used"] = data[TEMP_COL]
else:
    data["temp_used"] = (
        data.groupby("condition")[TEMP_COL]
            .rolling(window=SMOOTH_MEDIAN_WINDOW)
            .median()
            .reset_index(level=0, drop=True)
    )

# 2) Function: find the "first second" when temp_used >= THRESHOLD_C
def time_to_threshold(df_one_condition: "pd.DataFrame") -> float:
    hit = df_one_condition[df_one_condition["temp_used"] >= THRESHOLD_C]
    if hit.empty:
        return float("nan")  # threshold not reached
    return float(hit["seconds"].min())  # first second of crossing

# 3) Compute the metric for each condition
results = (
    data[data["condition"].isin(CONDITIONS)]
    .groupby("condition", as_index=False)
    .apply(lambda g: pd.Series({"time_to_threshold_s": time_to_threshold(g)}))
)

# So the order is desk/raised/fan
results["condition"] = pd.Categorical(results["condition"], categories=CONDITIONS, ordered=True)
results = results.sort_values("condition")

display(results)  # will show the table in Jupyter

# 4) Draw the bar chart
fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(results["condition"].astype(str), results["time_to_threshold_s"])
ax.set_title(f"Time-to-threshold: first time Temp ≥ {THRESHOLD_C}°C")
ax.set_xlabel("condition")
ax.set_ylabel("seconds")
plt.show()
```

```python
# ===== SETTINGS =====
TEMP_COL = "Temp:PackageId0,0"
FREQ_COL = "Frequency:Avg"
UTIL_COL = "Util:Avg"          # if your column is named differently — change it here
CONDITIONS = ["desk", "raised", "fan"]

# take only the needed columns + clean NaN
cols = ["condition", TEMP_COL, FREQ_COL, UTIL_COL]
df = cooling_data[cols].dropna()

# keep only 3 conditions
df = df[df["condition"].isin(CONDITIONS)]

# function: compute Pearson and Spearman for a pair of columns within one condition
def corr_pair(g, x, y):
    return {
        "pearson": g[x].corr(g[y], method="pearson"),
        "spearman": g[x].corr(g[y], method="spearman"),
        "n": len(g)
    }

rows = []
for cond in CONDITIONS:
    g = df[df["condition"] == cond]
    rows.append({
        "condition": cond,
        "Temp~Freq_pearson": corr_pair(g, TEMP_COL, FREQ_COL)["pearson"],
        "Temp~Freq_spearman": corr_pair(g, TEMP_COL, FREQ_COL)["spearman"],
        "Temp~Util_pearson": corr_pair(g, TEMP_COL, UTIL_COL)["pearson"],
        "Temp~Util_spearman": corr_pair(g, TEMP_COL, UTIL_COL)["spearman"],
        "n_points": len(g),
    })

corr_table = pd.DataFrame(rows)
display(corr_table)
```
